```
import pandas as pd
from sklearn.model_selection import train_test_split

# 1. Membaca Data Mengenai Movie dari IMDB
ImdbMovie_df = pd.read_csv('/content/MovieData-2018-2020.csv')

# Hapus baris yang kolom 'genre'-nya kosong (NaN)
ImdbMovie_df = ImdbMovie_df.dropna(subset=['genre'])

# Filter genre yang jumlahnya minimal 2
genre_counts = ImdbMovie_df['genre'].value_counts()
valid_genres = genre_counts[genre_counts >= 2].index
ImdbMovie_df = ImdbMovie_df[ImdbMovie_df['genre'].isin(valid_genres)]

# 2. Melakukan Stratified Sampling
_, stratified_sample = train_test_split(
    ImdbMovie_df,
    test_size=0.3,
    stratify=ImdbMovie_df['genre'],
    random_state=43
)

# 3. Mengambil 30 data teratas dari hasil sampling
sample_30_teratas = stratified_sample.head(30)

# 4. Menampilkan hasil sampling
print("Stratified Sampling (30 Data Teratas):")
display(sample_30_teratas[['name', 'genre', 'budget']])
```

```
import scipy.stats as stats
import math

# 1. Ambil kolom budget dari 30 data sampel teratas
data_budget = sample_30_teratas['budget']

# 2. Hitung statistik dasar
x_bar = data_budget.mean()
s = data_budget.std(ddof=1)
n = len(data_budget)

confidence_level = 0.95

# 3. Hitung nilai estimasi
alpha = (1 - confidence_level)
df = n - 1

standard_error = s / math.sqrt(n)
critical_t = stats.t.ppf(1 - alpha/2, df)
margin_of_error = critical_t * standard_error

# 4. Confidence Interval
lower_bound = x_bar - margin_of_error
upper_bound = x_bar + margin_of_error

# 5. Output
print("=== STATISTIK SAMPEL ===")
print("Jumlah Sampel (n)     :", n)
print("Rata-rata Sampel (x̄) :", round(x_bar, 2))
print("Standar Deviasi (s)   :", round(s, 2))

print("\n=== HASIL ESTIMASI ===")
print("Critical t-value      :", round(critical_t, 4))
print("Margin of Error       :", round(margin_of_error, 2))

print("\nConfidence Interval 95%")
print("Batas Bawah :", round(lower_bound, 2))
print("Batas Atas  :", round(upper_bound, 2))

print("\nKesimpulan:")
print(f"Estimasi rata-rata budget film populasi berada pada interval "
      f"{lower_bound:,.2f} sampai {upper_bound:,.2f}")
```
